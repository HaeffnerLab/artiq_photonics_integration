```python
import numpy as np
import matplotlib.pyplot as plt

# Load the 2D array from the text file
data = np.loadtxt('/home/lattice/artiq/artiq_working_dir/repository/image.txt', delimiter=',')

# Display the array using imshow
plt.imshow(data.reshape(-1,4).T, cmap='viridis', aspect='auto')

# Get current axis
ax = plt.gca()

# Set integer ticks for x-axis
ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

# Set integer ticks for y-axis
ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))

plt.colorbar(label="Value Scale")  # Add a color scale
plt.title("2D Data Visualization")
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,15)
plt.show()
```
