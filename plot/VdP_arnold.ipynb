```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Path to your .h5 file
file_path =  [
    '../results/2025-04-30/01/000073394-A7_Sync_VdP2_Px1x2.h5',
    '../results/2025-04-30/01/000073402-A7_Sync_VdP2_Px1x2.h5',
    '../results/2025-04-30/01/000073406-A7_Sync_VdP2_Px1x2.h5',
]

detune=[0, 100, 200]
res_all=[]
# Open the .h5 file in read mode
for i in range(len(file_path)):
    with h5py.File(file_path[i], 'r') as file:
        data = file['datasets']
        pmt_count=data['pmt_counts_avg_thresholded'][:]

        res=0.0

        for i in range(len(pmt_count)):
            res+=(pmt_count[i]-0.5)*np.exp(-1j*np.pi*i/len(pmt_count))

        res_all.append(np.abs(res))

plt.scatter(detune,res_all)
```
